<b>3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.</b>

```python
jml_record = df.size
jml_attribute = len(df.columns)
nama_attribute = df.columns.tolist()
jml_attribute_angka = len(df.select_dtypes('int').columns)

info = {
    "Jumlah record" : jml_record,
    "Jumlah attribute" : jml_attribute,
    "Nama attribute" : nama_attribute,
    "Jumlah attribute bertipe angka" : jml_attribute_angka
}

for key, value in info.items():
    print(key, ":", value)
```

* <b>jml_record = df.size</b> pernyataan yang menghitung jumlah total elemen dalam DataFrame df, yang setara dengan jumlah record dalam DataFrame.
* <b>jml_attribute = len(df.columns)</b> pernyataan yang menghitung jumlah kolom atau atribut dalam DataFrame df. dengan mengambil daftar kolom (atribut) dari DataFrame dan menghitung panjangnya menggunakan fungsi len.
* <b>nama_attribute = df.columns.tolist()</b> mengambil daftar nama kolom dalam DataFrame df dan mengonversinya menjadi daftar Python menggunakan metode tolist(). Hasilnya adalah daftar nama atribut.
* <b>jml_attribute_angka = len(df.select_dtypes('int').columns)</b> pernyataan yang menghitung jumlah kolom dalam DataFrame df yang memiliki tipe data 'int' (bilangan bulat). Pertama, df.select_dtypes('int') digunakan untuk memilih kolom-kolom dengan tipe data 'int', dan kemudian len digunakan untuk menghitung jumlah kolom yang telah dipilih.
* <b>info</b> sebuah kamus (dictionary) berisi informasi yang telah dihitung sebelumnya, seperti jumlah record, jumlah atribut, nama atribut, dan jumlah atribut dengan tipe data angka. Setiap informasi disimpan sebagai pasangan kunci-nilai dalam kamus.
* <b>for key, value in info.items()</b> loop for yang digunakan untuk mengiterasi melalui pasangan kunci-nilai dalam kamus info.
* <b>print(key, ":", value)</b> pernyataan print yang mencetak kunci (key) dan nilainya (value) untuk setiap pasangan kunci-nilai dalam kamus info, digunakan untuk mencetak informasi statistik ke layar.

<b>4. Tampilkan nilai dataset pada baris dimulai dengan 3 digit NIM akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice.<b/>

```python
df.iloc[13:, 4:]
```

* <b>iloc[]</b> atribut pengindeksan (indexing) pada DataFrame pandas yang digunakan untuk melakukan indexing berdasarkan posisi (integer).
* <b>[13:, 4:]</b> Dalam pengindeksan ini, ada dua bagian:
  1.  <b>13:</b> bagian indeks baris. Dengan menggunakan 13:, Anda memilih baris mulai dari baris dengan indeks 13 hingga akhir.
  2. <b>4:</b> bagian indeks kolom. Dengan menggunakan 4:, Anda memilih kolom mulai dari kolom dengan indeks 4 hingga akhir DataFrame.
