```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
df
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='status', data=df)
plt.title('Status Penempatan Mahasiswa')
plt.xlabel('Status')
plt.ylabel('Jumlah')
plt.show()
```

diagram batang yang menunjukkan jumlah penempatan mahasiswa berdasarkan <b>status</b> (mungkin 'Placed' dan 'Not Placed') dari DataFrame df. Diagram ini membantu Anda dengan cepat memahami distribusi penempatan mahasiswa dalam dataset tersebut.

```python
plt.figure(figsize=(5, 5))
df['hsc_s'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Persentase Spesialisasi Pendidikan Menengah')
plt.ylabel('')
plt.show()
```

Pie chart / diagram lingkaran diatas memberikan visualisasi persentase dari berbagai spesialisasi pendidikan menengah pada (kolom 'hsc_s') dalam dataset. Dengan menggunakan diagram lingkaran, Kita dapat dengan cepat melihat proporsi masing-masing spesialisasi pendidikan menengah dalam dataset, membantu kita memahami distribusi jenis-jenis spesialisasi.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='ssc_p', y='degree_p', data=df, marker='o')
plt.title('Hubungan antara Skor Sekolah Menengah dan Skor Kuliah')
plt.xlabel('Skor Sekolah Menengah (ssc_p)')
plt.ylabel('Skor Kuliah (degree_p)')
plt.show()
```

Line plot / diagram garis diatas memvisualisasikan hubungan antara skor sekolah menengah ('ssc_p') dan skor kuliah ('degree_p') dalam dataset. Titik-titik pada diagram menunjukkan hubungan antara dua variabel ini untuk setiap data yang ada, dengan simbol bulatan yang menandai titik-titik tersebut. Diagram garis digunakan untuk menunjukkan tren atau pola antara kedua variabel tersebut.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ssc_p', y='degree_p', hue='status', data=df)
plt.title('Hubungan antara Skor Sekolah Menengah dan Skor Kuliah (Berdasarkan Status Penempatan)')
plt.xlabel('Skor Sekolah Menengah (ssc_p)')
plt.ylabel('Skor Kuliah (degree_p)')
plt.show()
```

scatter plot diatas memvisualisasikan hubungan antara skor sekolah menengah dan skor kuliah, dengan mempertimbangkan status penempatan mahasiswa ('Placed' atau 'Not Placed'). Setiap titik dalam plot mewakili satu mahasiswa, dan warna titik tersebut menunjukkan status penempatan mahasiswa tersebut. Dengan melihat scatter plot ini, Anda dapat melihat apakah ada pola atau tren antara skor sekolah menengah, skor kuliah, dan apakah mahasiswa tersebut ditempatkan atau tidak.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='status', y='ssc_p', data=df)
plt.title('Distribusi Skor Sekolah Menengah (ssc_p) Berdasarkan Status Penempatan')
plt.xlabel('Status Penempatan')
plt.ylabel('Skor Sekolah Menengah (ssc_p)')
plt.show()
```

diagram box plot diatas kita dapat melihat distribusi skor sekolah menengah (ssc_p) untuk mahasiswa yang ditempatkan ('Placed') dan yang tidak ditempatkan ('Not Placed'). Diagram ini memberikan informasi tentang rata-rata, kuartil, nilai-nilai ekstrem, serta keberadaan outlier dalam skor sekolah menengah untuk setiap kategori status penempatan. Diagram box plot membantu kita untuk memahami sebaran data dan membandingkan distribusi skor sekolah menengah antara mahasiswa yang ditempatkan dan yang tidak ditempatkan.
