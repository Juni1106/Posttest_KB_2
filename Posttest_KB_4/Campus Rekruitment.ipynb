### One Hot Encoding

```python
ohe = OneHotEncoder(sparse_output=False)
oh_encoded = ohe.fit_transform(df_manipulation[['gender']])
```

menggunakan one-hot encoder untuk mengubah kolom 'gender' dari DataFrame menjadi representasi one-hot encoded, dan hasilnya disimpan dalam variabel "oh_encoded". Proses ini memungkinkan penggunaan informasi kategorikal dalam algoritma pembelajaran mesin yang membutuhkan input numerik.

```python
gender_encoded = pd.DataFrame(oh_encoded,
                      columns=['f', 'm'])
```

<b>"gender_encode"</b> DataFrame baru yang berisi hasil dari proses one-hot encoding untuk kolom 'gender', dengan dua kolom terpisah ('f' dan 'm'). DataFrame ini siap digunakan dalam analisis atau pelatihan model machine learning.

```python
gender_encoded
```

<b>gender_encoded</b> menunjukkan bahwa ini adalah variabel atau objek yang berisi hasil dari proses encoding (pengkodean) terhadap kolom 'gender' pada suatu dataset. Biasanya, dalam konteks machine learning, kolom 'gender' diubah menjadi representasi numerik agar dapat digunakan oleh algoritma pembelajaran mesin, seperti dari kata-kata 'male' dan 'female' menjadi angka-angka.

```python
df_manipulation = pd.concat([df_manipulation, gender_encoded], axis=1)
```

Setelah operasi pd.concat dilakukan, DataFrame df_manipulation diperbarui dengan menambahkan kolom-kolom (atau Series) dari gender_encoded. Dengan demikian, df_manipulation sekarang berisi data dari DataFrame awalnya dan juga kolom-kolom yang ada dalam gender_encoded. Penggabungan ini memungkinkan penggantian atau pengayaan data dalam df_manipulation dengan informasi dari gender_encoded. Hal ini berguna, misalnya, dalam mempersiapkan data untuk proses pembelajaran mesin, di mana kadang-kadang perlu menambahkan atau mengubah fitur-fitur sebelum melatih model.

```python
df_manipulation.drop(columns=['gender'], inplace=True)
```

drop(columns=['gender']): Menghapus kolom 'gender' dari DataFrame df_manipulation.
inplace=True: Menyatakan bahwa perubahan akan diterapkan langsung pada DataFrame df_manipulation yang ada, tanpa perlu membuat salinan DataFrame baru.
